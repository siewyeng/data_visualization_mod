```python
import pandas
import numpy
import matplotlib.pyplot as plt
```

```python
my_data = pandas.read_csv('politics and outlook.csv', low_memory = False )
```

```python
print(len(my_data))             #number of rows (observations)
print(len(my_data.columns))     #number of columns (variables)
```

```python
print("How interested are you in what's going on in government and politics? 1, extremely interested")
count1 = my_data['W1_A1'].value_counts().sort_index(ascending = True)   #sort = False to make sure values displayed in original order
#instead of by the count value
perc1 = my_data['W1_A1'].value_counts(normalize = True).sort_index(ascending = True)
print(count1)
```

```python
# alphas = [height/max(heights) for height in heights]
# colors = [(.1,.5,.6,.7,.8,a) for a in alphas]
perc1.plot.bar()
plt.title("How interested are you in what's going on in government and politics?", fontsize=16)
plt.ylabel('Proportion responses',fontsize=12)
plt.xlabel('1 for "extremely interested"', fontsize = 12)
plt.xticks(rotation = 0, fontsize=12)
plt.show()
```

```python
print("Do you remember whether or not you voted in that election? 1, yes")
count2 = my_data['W1_A4'].value_counts().sort_index(ascending = True)
perc2 = my_data['W1_A4'].value_counts(sort = False, normalize = True).sort_index(ascending = True)
print(count2)

perc2.plot.bar()
plt.title("Do you remember whether or not you voted in that election?", fontsize=16)
plt.ylabel('Proportion responses',fontsize=12)
plt.xlabel('1 for "yes"', fontsize = 12)
plt.xticks(rotation = 0, fontsize=12)
plt.show()
```

```python
print("How far along the road to your American Dream do you think you will ultimately get on a 10-point scale where 1 is not far at all and 10 nearly there?")
count6 = my_data['W1_F6'].value_counts(sort = False).sort_index(ascending = True)
perc6 = my_data['W1_F6'].value_counts(sort = False, normalize = True).sort_index(ascending = True)
print(count6)

perc6.plot.bar()
plt.title("How far along the road to your American Dream do you think you will ultimately get on a 10-point scale?", fontsize=16)
plt.ylabel('Proportion responses',fontsize=12)
plt.xlabel('1 is not far at all and 10 nearly there', fontsize = 12)
plt.xticks(rotation = 0, fontsize=12)
plt.show()
```
